```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cn2 = pd.read_csv('chungnam2.csv')
```

```python
cn2.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
cn2.head()
```

```python
def town(add):
  splitted = add.split()
  if '충남' in splitted[0]:
    if '천안시' in splitted[1]:
      return splitted[3]
    else:
      return splitted[2]
  elif '천안' in splitted[0]:
    return splitted[2]
  elif '충청남도' in splitted[0]:
    if len(splitted) == 3:
      return splitted[2]
    else:
      if '천안시' in splitted[1]:
        if len(splitted) >4:
          return splitted[3]
        else:
          if '구' in splitted[2]:
            return splitted[3]
          else:
            return splitted[2]
      else:
        return splitted[2]
  else:
    return 'unknown'

cn2['town'] = cn2.apply(lambda x: town(x['HPPN_PNU_ADDR']), axis=1)
cn2.head()
```

```python
cn2.town.value_counts()
```

```python
def city(add):
  splitted = add.split()
  if len(splitted) > 1:
    if '충남' in splitted[0]:
      return splitted[1]
    elif '충청' in splitted[1]:
      return splitted[2]
    else:
      return splitted[1]
  else:
    return "unknown"

cn2['city'] = cn2.apply(lambda x: city(x['HPPN_PNU_ADDR']), axis=1)
cn2.head()
```

```python
cn2.city.value_counts()
```

```python
def above_three(town):
  return len(town)>3
cn2['is_town'] = cn2.apply(lambda x: above_three(x['town2']), axis=1)
cn2['is_town'].value_counts()
```

```python
def count_word(town):
  return len(town)
cn2['word_town'] = cn2.apply(lambda x: count_word(x['town2']), axis=1)
cn2['word_town'].value_counts()
```
