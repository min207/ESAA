#### **- matplotlib 라이브러리 자유자재로 사용하기**

- seaborn 라이브러리의 tips라는 데이터 사용
  tips: 식당에서 팁을 지불한 손님의 정보 데이터

```python
tips = sns.load_dataset("tips")     # 데이터 불러오기
print(tips.head())
print(type(tips))       
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)      # 기본 틀, 격자 구성
```

- 히스토그램
  + 데이터프레임의 열 데이터 분포와 빈도를 살펴보는 용도로 주로 사용
  + 변수를 하나만 사용 > 일변량 그래프

```python
axes1.hist(tips['total_bill'], bins=10)       # 히스토그램 생성 / bins: x축의 간격 조정
axes1.set_title('Histogram of Total Bill')    # 그래프 제목 생성
axes1.set_xlabel('Frequency')                 # x축 제목 생성
axes1.set_ylabel('Total Bill')                # y출 제목 생성

fig
```

- 산점도 그래프
  + 하나의 변수에 따른 다른 변수의 분포를 알 수 있음
  + 변수 2개 사용 > 이변량 그래프


```python
scatter_plot = plt.figure()

axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

- 박스 그래프
  + 이산형 변수와 연속형 변수를 함께 사용
    - 이산형 변수: 명확하게 구분되는 값
    - 연속형 변수 : 명확하게 셀 수 없는 범위의 값

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex'] == 'Female']['tip'],
               tips[tips['sex'] == 'Male']['tip']],
              labels = ['Female', 'Male'])
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

- 다변량 그래프
  + 3개 이상의 변수를 사용한 그래프
  + 점의 색상을 다르게 하는 등 다양한 방법을 통해 시각화 가능

```python
def recode_sex(sex):                # 문자열을 정수로 치환
  if sex == 'Female':
    return 0
  else:
    return 1

tips['sex_color'] = tips['sex'].apply(recode_sex)       # sex_color 열 추가

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10,
    c=tips['sex_color'],
    alpha=0.5
)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

#### **- seaborn 라이브러리 자유자재로 사용하기**

```python
import seaborn as sns

tips = sns.load_dataset("tips")   # 데이터 불러오기
```

```python
ax = plt.subplots()                         # 기본 틀
ax = sns.distplot(tips['total_bill'])       # 히스토그램
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde = False)    # kde = False로 설정해 밀집도 그래프 제외
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist = False)   # hist = False를 이용해 히스토그램 제외
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True)     # 양탄자 그래프 추가
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

```python
ax = plt.subplots()
ax = sns.countplot('day', data=tips)    # 카운트 그래프: 이산값을 나타냄
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

- regplot 메서드 > 산점도 그래프
  + 회귀선도 함께 그림
  + fit_reg = False를 설정해 회귀선 제외 가능

```python
ax = plt.subplots() 
ax = sns.regplot(x='total_bill', y='tip', data=tips) 
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots() 
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)     # 회귀선 제거
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip')
```

- jointplot 메서드 > 산점도+히스토그램
  + 이때, kind = "hex"를 설정하여 산점도 대신 육각 그래프 사용 가능

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips)   
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex") 
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

- kdeplot 메서드 > 이차원 밀집도
  + shade = True를 설정해 그래프에 음영 효과 부여

```python
kde, ax = plt.subplots() 
ax = sns.kdeplot(data=tips['total_bill'],
                 data2=tips['tip'], 
                 shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip')
```

- barplot 메서드 > 바 그래프

- boxplot 메서드 > 박스 그래프

```python
ax = plt.subplots() 
ax = sns.violinplot(x='time', y='total_bill', data=tips) 
ax.set_title('Violin plot of total bill by time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Total Bill')
```

- pairplot 메서드 > 관계 그래프
  + 중복된 정보가 표현된다는 단점이 있음 >
위치를 직접 지정하여 원하는 그래프로 교체 가능

```python
fig = sns.pairplot(tips)
```

```python
pair_grid = sns.PairGrid(tips) 
pair_grid = pair_grid.map_upper(sns.regplot)            # 대각선 기준 위쪽 그래프
pair_grid = pair_grid.map_lower(sns.kdeplot)            # 대각선 기준 아래쪽 그래프
pair_grid = pair_grid.map_diag(sns.distplot, rug=True)  # 대각선 중심 그래프
plt.show()
```

#### **- seaborn 라이브러리로 그래프 스타일 설정하기**
- set_style 메서드 사용
  + 한 번 실행 시 이후 모든 그래프에 적용

```python
fig, ax=plt.subplots()
ax=sns.violinplot(x='time',y='total_bill',hue='sex',data=tips,split=True)
```

```python
sns.set_style('whitegrid') # 그래프 배경에 가로줄 생성
fig, ax=plt.subplots()
ax=sns.violinplot(x='time',y='total_bill',hue='sex',data=tips,split=True)
```

```python
fig=plt.figure()

seaborn_styles=['darkgrid','whitegrid','dark','white','ticks']

for idx, style in enumerate(seaborn_styles):
  plot_position = idx+1
  with sns.axes_style(style):
    ax = fig.add_subplot(2,3,plot_position)
    violin = sns.violinplot(x='time',y='total_bill',data=tips, ax=ax)
    violin.set_title(style)

fig.tight_layout()
```
